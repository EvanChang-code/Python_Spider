```python
import pymongo
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
company_info = client['company_info']
information = company_info['information_1']
```

```python
for i in information.find():
    print(i)
```

```python
# 通过聚合管道来晒选数据，数据测试使用
pipeline = [
#     {'$match':{'date':'2016.11.28'}},
    {'$group':{'_id':'$industry','counts':{'$sum':1}}},
    {'$sort':{'counts':-1}},
]
```

```python
for i in information.aggregate(pipeline): # 数据测试使用
    print(i)
```

```python
def data_gen():  # 数据类目生成器
    pipeline = [
    {'$group':{'_id':'$industry','counts':{'$sum':1}}},
    {'$sort':{'counts':-1}},
]
    for i in information.aggregate(pipeline):
        yield [i['_id'],i['counts']]
```

```python
# 固定格式，在饼图中需要注意的数据为列表，应注意不用的类目对应不同的数据
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': '不同服务类型公司融资数量'},
    'subtitle': {'text': '不同服务类型公司融资数量分布占比'},
    }
series = {
    'name':'pie charts',
    'data':[i for i in data_gen()],
    'type':'pie'
}
charts.plot(series,options=options,show='inline')
```
